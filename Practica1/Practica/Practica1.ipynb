```python
from utils import load_data 
from linear_reg import gradient_descent
from linear_reg import compute_gradient
from linear_reg import compute_cost
from public_tests import compute_cost_test
from public_tests import compute_gradient_test
from public_tests import compute_gradient_descent_test

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

# Carga los datos desde el archivo
X, y = load_data()

# Grafica los puntos en el plano XY
plt.scatter(X, y, marker='o', color='blue', label='Puntos')

coefficient = np.polyfit(X, y, 1);
m, b = coefficient

X_values = np.linspace(min(X), max(X), 100)  # Valores de X para el gráfico
Y_Values = m * X_values + b;
plt.plot(X_values, Y_Values, color='red', label='Recta de Regresión')

plt.grid(True)  # Muestra una cuadrícula en la gráfica

# Muestra la gráfica
plt.show()

compute_cost_test(compute_cost)

compute_gradient_test(compute_gradient)

w, b, Jhistory = gradient_descent(X, y, 0, 0, compute_cost, compute_gradient, 0.01, 1500)

compute_gradient_descent_test(w, b)




# Crea un arreglo con los números de iteración
iterations = np.arange(1, 1500 + 1)

# Grafica la curva de aprendizaje (historial de costos)
plt.plot(iterations, Jhistory, color='blue')

plt.show();
# Grafica los puntos en el plano XY
plt.scatter(X, y, marker='o', color='blue', label='Puntos')
y_pred = w * X + b
plt.plot(X, y_pred, color='red', label='Recta de Regresión')
# Personaliza el gráfico si es necesario
plt.title('Recta de Regresión Lineal')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.grid(True)

# Muestra el gráfico
plt.show()
```
