```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

x = np.loadtxt("data/AllData.csv", delimiter=',', usecols = (0, 1, 2, 3, 4, 5, 6, 7, 8))


# delimiter specifies the character separating values in your CSV (e.g., ',', ';', '\t' for tab-delimited)
# dtype=None allows NumPy to infer the data type of each column
# skip_header skips the specified number of lines at the beginning of the file (adjust as needed)
y = np.genfromtxt("data/AllData.csv", delimiter=',', dtype=None, encoding='utf-8', usecols = 9)

# 'names=True' assumes that the first row contains column names
# 'encoding' parameter is optional but can be useful for handling different text encodings

# Accessing columns by their names

plt.scatter(x[:,3], y, color='blue')
plt.show()
```
