```python
from logistic_reg import sigmoid
from logistic_reg import compute_cost
from logistic_reg import compute_gradient
from logistic_reg import compute_cost_reg
from logistic_reg import compute_gradient_reg
from logistic_reg import gradient_descent
from logistic_reg import predict

from public_tests import sigmoid_test
from public_tests import compute_cost_test
from public_tests import compute_gradient_test
from public_tests import compute_cost_reg_test
from public_tests import compute_gradient_reg_test
from public_tests import predict_test

from utils import load_data_multi_data1
from utils import load_data_multi_data2
from utils import load_data_multi_data2_separatedValues
from utils import plot_data
from utils import plot_decision_boundary
from utils import map_feature

import matplotlib.pyplot as plt
import numpy as np


############## PART A ##############

# Cargar Datos
x, y = load_data_multi_data1()

# Mostrar Datos
plot_data(x, y)
plt.show()

########### TEST ##########
sigmoid_test(sigmoid)
compute_cost_test(compute_cost)
compute_gradient_test(compute_gradient)


n = x.shape[1]
w_init = np.zeros(n)
w,b,JHistory = gradient_descent(x,y,w_init,-8,compute_cost,compute_gradient,0.001,10000)

print(JHistory[-1])

print("w = ")
print(w)

plot_decision_boundary(w, b, x, y)


p = predict(x, w, b)
predict_test(predict)
plot_decision_boundary(w, b, x, p)
plt.show()


############## PART B ##############


# Cargar Datos
x_data2, y_data2 = load_data_multi_data2()

# Mostrar Datos
plot_data(x_data2, y_data2)
plt.show()


########### TEST ##########
compute_cost_reg_test(compute_cost_reg)
compute_gradient_reg_test(compute_gradient_reg)


########### GRADIENT DESCENT ##########
x1, x2, y = load_data_multi_data2_separatedValues()
Xmap = map_feature(x1, x2)

w,b,Jhistory = gradient_descent(Xmap,y_data2,np.zeros(Xmap.shape[1]),1,compute_cost_reg,compute_gradient_reg,0.01,10000,0.01)
plot_decision_boundary(w,b,Xmap,y_data2)

########### PREDICTION ##########
pred = predict(Xmap, w, b)
percent = np.mean(y == pred)
print("Prediction Percent = ", percent *100, " % ")
```
